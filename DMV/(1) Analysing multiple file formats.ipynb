# Load the sales data from each file format into the appropriate data structures or Dataframes.

```python
import pandas as pd

customers_csv = pd.read_csv('../../Datasets/customers.csv')
customers_json = pd.read_json('../../Datasets/customers.json')
customers_excel = pd.read_excel('../../Datasets/customers.xlsx')
```

```python
customers_csv['registered'] = pd.to_datetime(customers_csv['registered'])
customers_json['registered'] = pd.to_datetime(customers_json['registered'])
```

## Checking for null values

```python
customers_csv.isna().sum()
```

```python
customers_excel.isna().sum()
```

```python
customers_json.isna().sum()
```

```python
customers_csv.fillna(0,inplace=True)
customers_json.fillna(0,inplace=True)
customers_excel.fillna(0,inplace=True)
```

## checking for duplicates 

```python
customers_csv[customers_csv.duplicated()]
```

```python
customers_json[customers_json.duplicated()]
```

```python
customers_excel[customers_excel.duplicated()]
```

```python
customers_csv = customers_csv.drop_duplicates()
customers_json = customers_json.drop_duplicates()
customers_excel = customers_excel.drop_duplicates()
```

# Convert the data into a unified format, such as a common dataframe or data structure, to enable seamless analysis.

```python
uni_df = pd.concat([customers_csv,customers_excel,customers_json],ignore_index=True)
uni_df.head()
```

# Perform data transformation tasks, such as merging multiple datasets, splitting columns, or deriving new variables.

## deriving new variable by using two variable

```python
uni_df['full name'] = uni_df['first_name'] + ' ' + uni_df['last_name']
```

```python
uni_df.head()
```

## Deriving new variable from one variable

```python
def extract_pin_code(address):
    pin_code = [single_add[-5:] for single_add in address]
    return pin_code

uni_df['pin code'] = extract_pin_code(uni_df['address'])
```

```python
uni_df.sample(3)
```

# Analyse the sales data by performing descriptive statistics, aggregating data by specific variables, or calculating metrics such as total sales, average order value, or product category distribution.

```python
uni_df.describe()
```

```python
uni_df.groupby('job').agg({'spent':'sum','orders':'mean'})
```

```python
uni_df['spent'].sum()
```

```python
uni_df['orders'].mean()
```

```python
uni_df['job'].value_counts()
```

# Create visualisations, such as bar plots, pie charts, or box plots, to represent the sales data and gain insights into sales trends, customer behaviour, or product performance.

```python
uni_df.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns



plt.figure(figsize=(10,5))
sns.barplot(x='job',y='spent',data=uni_df)
plt.xticks(rotation=90);
```

```python
job_valcount = uni_df['job'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(x=job_valcount,labels=job_valcount.index,autopct='%1.2f%%',rotatelabels=True,pctdistance=0.9);
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x='job',y='spent',data=uni_df)
plt.xticks(rotation=90);
```
